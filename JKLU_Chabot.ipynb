```python



# ---------------------------------------------------
# JKLU Open-RAG Chatbot — FINAL WORKING VERSION
# Excel Priority + RAG + Strict PII + Stable UI
# ---------------------------------------------------

import os
import re
import requests
import warnings
import pandas as pd
import gradio as gr

warnings.filterwarnings("ignore")
os.environ["TOKENIZERS_PARALLELISM"] = "true"


# ---------------------------------------------------
GOOGLE_API_KEY = "AIzaSyC22sj1EeatqswKCDYhD4FDFvnxNbTqu_k"
SEARCH_ENGINE_ID = "712b04288572145e4"

os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
os.environ["SEARCH_ENGINE_ID"] = SEARCH_ENGINE_ID


# ---------------------------------------------------
# ENABLE GEMINI
# ---------------------------------------------------
try:
    import google.generativeai as genai
    genai.configure(api_key=GOOGLE_API_KEY)
    MODEL_CLIENT = genai.GenerativeModel("gemini-2.5-flash")
except:
    MODEL_CLIENT = None


# ---------------------------------------------------
# LOAD STUDENT EXCEL
# ---------------------------------------------------
STUDENT_EXCEL_FINAL = "/content/cleaned_marks.xlsx"

if not os.path.exists(STUDENT_EXCEL_FINAL):
    raise FileNotFoundError("Upload cleaned_marks.xlsx to /content.")

STUDENTS = pd.read_excel(STUDENT_EXCEL_FINAL)
STUDENTS["Total"] = pd.to_numeric(STUDENTS["Total"], errors="coerce")

if "Roll" not in STUDENTS.columns:
    for c in STUDENTS.columns:
        if "roll" in c.lower():
            STUDENTS.rename(columns={c: "Roll"}, inplace=True)


# ---------------------------------------------------
# PII SYSTEM
# ---------------------------------------------------
PERSONAL_KEYS = ["email", "phone", "contact", "address", "mobile",
                 "mother", "father", "dob", "name", "student name"]

def asks_personal_info(msg: str) -> bool:
    msg = msg.lower()
    return any(word in msg for word in PERSONAL_KEYS)


def mask_pii(text: str) -> str:
    if not isinstance(text, str):
        return text
    text = re.sub(r'([A-Za-z0-9_.+-]{1})[A-Za-z0-9_.+-]*@', r'\1***@', text)
    text = re.sub(r'\b\d{10}\b', "**********", text)
    return text


# ---------------------------------------------------
# EXCEL FUNCTIONS
# ---------------------------------------------------
def student_summary():
    df = STUDENTS
    total = len(df)
    avg = df["Total"].mean()
    passed = (df["Total"] >= 30).sum()
    failed = (df["Total"] < 30).sum()
    grades = df["Grade"].value_counts().to_dict()
    return f"Total: {total}, Average: {avg:.2f}, Passed: {passed}, Failed: {failed}, Grades: {grades}"

def count_grade(g):
    df = STUDENTS
    return f"Number of students with grade {g}: {(df['Grade'].astype(str).str.upper() == g).sum()}"

def student_top_k(k=5):
    df = STUDENTS.sort_values("Total", ascending=False).head(k)
    resp = []
    for i, (_, r) in enumerate(df.iterrows(), 1):
        resp.append(f"Rank {i}: Total = {r['Total']}")
    return "\n".join(resp)

def student_by_roll(roll):
    df = STUDENTS
    row = df[df["Roll"].astype(str).str.contains(str(roll))]
    if row.empty:
        return "No student found."
    r = row.iloc[0]
    return f"Name: [REDACTED]\nRoll: {str(r['Roll'])[:4]}***\nTotal: {r['Total']}\nGrade: {r['Grade']}"


# ---------------------------------------------------
# GOOGLE SEARCH
# ---------------------------------------------------
def google_search(q, num=5):
    try:
        r = requests.get(
            "https://www.googleapis.com/customsearch/v1",
            params={"q": q, "cx": SEARCH_ENGINE_ID, "key": GOOGLE_API_KEY, "num": num},
            timeout=5
        )
        return r.json().get("items", [])
    except:
        return []


def clean_snips(snips):
    out = []
    for s in snips[:3]:
        sn = s.get("snippet", "").replace("...", "")
        if sn:
            out.append(sn)
    return " ".join(out)


def call_gemini(prompt):
    try:
        res = MODEL_CLIENT.generate_content(prompt)
        return res.text
    except:
        return ""


# ---------------------------------------------------
# SCOPE GUARD
# ---------------------------------------------------
SCOPE = [
    "jklu", "where", "location", "located", "campus",
    "institute", "university", "program", "course", "department",
    "admission", "specialization", "design", "result", "top",
    "grade", "marks", "highest", "average", "roll", "summary",
    "performance", "class", "fees", "hostel"
]

def input_allowed(msg):
    msg = msg.lower()
    return any(s in msg for s in SCOPE)


# ---------------------------------------------------
# MAIN CHAT RESPONSE
# ---------------------------------------------------
def chat_response(user, history=None):
    msg = user.lower()
    history = history or []

    # PII FIRST
    if asks_personal_info(msg):
        return history + [(user, "Sorry, I cannot provide personal student information.")]

    # Grade Query
    gmatch = re.search(r"(grade\s*[a-f])|([a-f]\s*grade)|(\bgrade\b.*?[a-f])", msg, flags=re.I)
    if gmatch:
        g = re.search(r"[A-Fa-f]", gmatch.group(0)).group(0).upper()
        return history + [(user, count_grade(g))]

    # Summary
    if any(w in msg for w in ["summary", "average", "performance", "pass", "fail", "class"]):
        return history + [(user, student_summary())]

    # Top students
    if "top" in msg or "highest" in msg:
        return history + [(user, student_top_k())]

    # Roll search
    roll = re.search(r"\b([0-9]{4}[A-Za-z0-9_-]{2,12}[0-9]{1,6})\b", msg)
    if roll:
        return history + [(user, student_by_roll(roll.group(1)))]

    # Scope Guard
    if not input_allowed(msg):
        return history + [(user, "Only JKLU-related queries allowed.")]

    # GOOGLE RAG
    query = f"JK Lakshmipat University {user}"
    snips = google_search(query)

    if not snips:
        return history + [(user, "Information not found.")]

    merged = "\n".join(
        (s.get("title", "") + ". " + s.get("snippet", "")).replace("...", "")
        for s in snips
    )

    prompt = f"""
Use ONLY this context:
{merged}

Question:
{user}

If answer missing → say "Information not found."
"""

    ans = call_gemini(prompt) or clean_snips(snips)
    ans = mask_pii(ans)

    return history + [(user, ans)]



# BEAUTIFUL UI
# ---------------------------------------------------
def launch_ui():
    custom_css = """
    body {
        background: linear-gradient(135deg, #ffd39f, #ff9e6f);
        background-attachment: fixed;
    }

    .container {
        max-width: 1250px;
        margin: auto;
    }

    .header-box {
        padding: 20px;
        border-radius: 20px;
        background: rgba(255, 255, 255, 0.55);
        backdrop-filter: blur(10px);
        box-shadow: 0px 6px 25px rgba(0,0,0,0.15);
        margin-bottom: 10px;
    }

    .sidebar-box {
        padding: 20px;
        background: rgba(255, 255, 255, 0.55);
        border-radius: 20px;
        box-shadow: 0px 6px 20px rgba(0,0,0,0.12);
        backdrop-filter: blur(10px);
    }

    .chatbox {
        padding: 12px;
        border-radius: 20px;
        background: rgba(255, 255, 255, 0.45);
        box-shadow: 0px 6px 20px rgba(0,0,0,0.12);
        backdrop-filter: blur(12px);
        height: 550px;
    }

    .input-bar {
        background: rgba(255, 255, 255, 0.65);
        backdrop-filter: blur(8px);
        padding: 12px;
        border-radius: 15px;
        box-shadow: 0px -2px 15px rgba(0,0,0,0.08);
        margin-top: 10px;
    }
    """

    with gr.Blocks(theme=gr.themes.Soft(primary_hue="orange"), css=custom_css) as demo:

        # -------- HEADER --------
        gr.HTML("""
        <div class="container">
            <div class="header-box">
                <h1 style="text-align:center; font-size:36px; margin-bottom:5px;">
                    🎓 JKLU Smart AI Chatbot
                </h1>
                <p style="text-align:center; color:#444; font-size:17px; margin-top:-8px;">
                    Excel • RAG • Gemini • Privacy Protected • Instant Answers
                </p>
            </div>
        </div>
        """)

        with gr.Row(elem_classes="container"):

            # -------- SIDEBAR --------
            with gr.Column(scale=1, min_width=270):
                gr.HTML("""
                <div class="sidebar-box">
                    <h3 style="margin-top:0;">📘 System Overview</h3>
                    <ul>
                        <li>Excel Loaded</li>
                        <li>PII Protection ON</li>
                        <li>Google RAG Active</li>
                        <li>Gemini Model Ready</li>
                    </ul>

                    <h3>🧠 Bot Features</h3>
                    <ul>
                        <li>Marks & Grade Stats</li>
                        <li>Class Summary</li>
                        <li>Top 5 Students</li>
                        <li>Roll Number Lookup</li>
                        <li>JKLU Admission Info</li>
                        <li>Campus & Programs Details</li>
                    </ul>

                    <h3>🔒 Safety Rules</h3>
                    <ul>
                        <li>Emails Masked</li>
                        <li>Phone Numbers Blocked</li>
                        <li>Roll Numbers Hidden</li>
                        <li>No Personal Data Shared</li>
                    </ul>
                </div>
                """)

            # -------- CHAT AREA --------
            with gr.Column(scale=4):

                chatbot = gr.Chatbot(
                    label="JKLU Assistant",
                    height=550,
                    elem_classes="chatbox"
                )

                with gr.Row(elem_classes="input-bar"):
                    msg = gr.Textbox(
                        placeholder="Ask anything about JKLU, grades, admissions, results...",
                        scale=8,
                        show_label=False
                    )
                    send = gr.Button("Send", variant="primary", scale=1)
                    clear = gr.Button("Clear", variant="secondary", scale=1)

                # Message Handling Logic
                def add_user_message(m, history):
                    if not m:
                        return "", history
                    return "", history + [(m, None)]

                def bot_reply(history):
                    user_msg = history[-1][0]
                    return chat_response(user_msg, history[:-1])

                # Events
                msg.submit(add_user_message, [msg, chatbot], [msg, chatbot]) \
                   .then(bot_reply, chatbot, chatbot)

                send.click(add_user_message, [msg, chatbot], [msg, chatbot]) \
                    .then(bot_reply, chatbot, chatbot)

                clear.click(lambda: [], None, chatbot)

    return demo





# ---------------------------------------------------
# RUN APP
# ---------------------------------------------------
if __name__ == "__main__":
    print("Loaded:", STUDENT_EXCEL_FINAL)
    ui = launch_ui()
    ui.launch(debug=True)
```
